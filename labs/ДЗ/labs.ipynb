```python
students.index.name = 'ID'
students = students.dropna(how='all', axis=1)
students = students.astype('float')
students
```

```python
attendance = students.iloc[:, students.columns.get_level_values(2) == 'посещ-е'].sum(axis=1)
active = students.iloc[:, students.columns.get_level_values(2) == 'актив'].sum(axis=1)
lessonsAmount = len(students.iloc[:, students.columns.get_level_values(2) == 'посещ-е'].columns)
activeAmount = len(students.iloc[:, students.columns.get_level_values(2) == 'актив'].columns)
```

```python
file1 = pd.read_excel('1 - Степень с натуральным и целым показателем.xlsx') 
file2 = pd.read_excel("2 - Степень с рациональным показателем.xlsx") 
file3 = pd.read_excel("3 - Многочлен Разложение на множители.xlsx") 
file4 = pd.read_excel("Успеваемость групп.xlsx",sheet_name='ДЗ',header=None) 
file5 = pd.read_excel("5 - Линейные и квадратные уравнения.xlsx") 
file6 = pd.read_excel("6 - Линейные неравенства и их системы.xlsx")
```

```python
def fileParce(fileName, renameS):
    fileName = fileName[['Процент правильных ответов (%)', 'ID']]
    fileName = fileName.dropna(axis=0, how='any')
    fileName = fileName.set_index('ID')
    fileName = fileName.rename(columns={
        'Процент правильных ответов (%)': renameS
    })
    fileName = fileName.groupby('ID', as_index=True).max()
    return fileName
```

```python
file1 = fileParce(file1,'Степени с натур. и целым показателем')
file2 = fileParce(file2,'Степени с рацион-м показателем')
file3 = fileParce(file3,'Многочлены. Разложение на множители')
file5 = fileParce(file5,'Линейные и квадратные уравнения')
file6 = fileParce(file6,'Линейные неравенства и их системы')
```

```python
kt = pd.read_excel("Успеваемость групп.xlsx",sheet_name='КТ', header=[1])
kt.drop(kt.columns[[5,6,7]], axis = 1, inplace = True)
kt = kt.dropna(how='all')
```

```python
kt = kt.replace('н',0)
kt['Решение неравенств и их систем'] = kt['Решение неравенств и их систем'].astype(str)
kt['Решение неравенств и их систем'] = kt['Решение неравенств и их систем'].str.replace(',','.')
kt = kt.fillna(0)
kt[['Преобразование выражений','Решение уравнений и их систем','Решение неравенств и их систем']] = kt[['Преобразование выражений','Решение уравнений и их систем','Решение неравенств и их систем']].astype(float)
kt['Итого'] = kt.loc[:,['Преобразование выражений','Решение уравнений и их систем','Решение неравенств и их систем']].sum(axis=1)
kt = kt.set_index('Тема контрольной  точки')
kt.index.name = 'ID'
kt
```

```python
attend = attendance/lessonsAmount*100
acti = active/activeAmount*100
ktCount = kt['Итого']/25*100
totalSumBal = 0.1*(0.7*attend+0.3*acti)+0.2*fileCount+0.7*ktCount
final_table = pd.concat([attendance ,active, attend, acti, fileCount, ktCount,totalSumBal], axis=1)
final_table = final_table.drop('ID')
```

```python
def rating(row):
    if row < 50:
        return 'Н/A'
    elif 50<=row<=69:
        return '3'
    elif 70<=row<=84:
        return '4'
    else:
        return '5'
```

```python

final_table['bal'] = final_table[5].apply(rating)
final_table = final_table.rename(columns={
        0: 'Посещение',
        1: 'Активность',
        2: '% посещения',
        3: '% активности',
        4: 'ДЗ',
        'Итого': 'КТ',
        5: 'ИТОГ'
})
```
